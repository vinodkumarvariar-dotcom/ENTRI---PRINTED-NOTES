### transpose of a high dimensional array with specified order of axes

```python
2 blocks (or outermost arrays)

Each block has 2 rows

Each row has 4 columns
```

```python
arr1 = np.arange(16).reshape((2,2,4))
arr1
```

```python
arr1.ndim
```

```python
arr1 = np.arange(16).reshape((2,2,4))
print(arr1)

# #  first two axes swapped.
arr2 = arr1.transpose((1,0,2))
print(arr2)
```

### swap axes

```python
arr1 = np.arange(16).reshape((2,2,4))
print(arr1)
print(arr1.swapaxes(1,2))
```

### append elements to an array

```python
arr = np.array([1,2,3])
```

```python
# append a scalar and return a copy
arr1 = np.append(arr, 4)    
print(arr1)
```

```python
# append an array and return a copy
arr2 = np.append(arr1, [4,5,6])    
print(arr2)
```

### insert elements into an array

```python
# np.insert(array, position, element)

# insert a scalar at a certain position
arr3 = np.insert(arr, 0, 100)    
print(arr3)
```

```python
# insert multiple values at a certain position
arr3 = np.insert(arr, 0, [5,7,8])    
print(arr3)
```

### delete elements from an array

```python
# remove the element at position 0
arr4 = np.delete(arr, 0)    
print(arr4)
```

```python
# remove the element at multiple positions
arr4 = np.delete(arr, [0,2])    
print(arr4)
```

### copy an array

```python
arr = np.array([1,2,3])
```

```python
# the following methods are all copy
arr1 = np.copy(arr)
# or 
arr1 = arr.copy()
# or 
arr1 = np.array(arr, copy=True)
```
