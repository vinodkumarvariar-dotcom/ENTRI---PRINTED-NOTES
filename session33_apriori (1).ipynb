```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data = {
    'Bread':[1,1,1,0,1],
    'Butter':[1,1,0,1,1],
    'Milk':[1,0,1,0,0],
    'Jam':[0,0,0,1,1]
}
df = pd.DataFrame(data)
df
```

```python
frequent_itemset = apriori(df,min_support=0.4,use_colnames=True)
frequent_itemset
```

```python
# use_colnames True means the output will show item names instead of numbers
```

```python
rules = association_rules(frequent_itemset,metric='confidence',min_threshold=0.6)
rules[['antecedents','consequents','support','confidence','lift']]
```

```python
# confidence - to focus on how often the rule is true
# antecedents - the if part of rule
# consequents - the then part of the rule
# support - how often the itmes in the rule appear together
# lift -- how much more likely the items are bought together compared to seperately(lift>1 means a strong pattern)
```


Bread and Butter are strongly linked:

Rule: "If Bread, then Butter" has 60% support (they appear together in 3 lists) and 75% confidence (75% of people who buy Bread also buy Butter).

The reverse rule ("If Butter, then Bread") is also true with 75% confidence.

Insight: Customers who buy Bread are very likely to buy Butter, and vice versa. A store could bundle these items or place them close together.

Note: The lift (0.9375) is slightly less than 1, suggesting the connection is strong but not much stronger than if Bread and Butter were bought independently (since both are very common). It means Are people really buying two items together because they are connected, or is it just by chance?"

Lift > 1 → People are buying them together more than expected → There’s a real connection.

Lift = 1 → Buying together is just coincidence.

Lift < 1 → Buying together happens less than expected → Probably no real connection.
