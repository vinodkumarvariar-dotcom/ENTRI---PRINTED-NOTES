```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
texts = [
    "I love this movie",
    "This film is amazing",
    "I hate this movie",
    "This movie is terrible",
    "Absolutely fantastic experience",
    "Worst movie ever"
]

labels = [1, 1, 0, 0, 1, 0]  # 1 = positive, 0 = negative
```

num_words = vocab_size

Keeps only top 1000 frequent words

oov_token="<OOV>"

Handles unknown words

If a new word appears → replaced with <OOV>

```python
vocab_size = 1000
max_length = 10   #Every sentence will be converted into a sequence of exactly 10 numbers.fixed no of inputs

tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)   # tokenizer.word_index
# {'<OOV>':1, 'this':2, 'movie':3, 'love':4}

sequences = tokenizer.texts_to_sequences(texts)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

labels = tf.constant(labels)
```

```python
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=64, input_length=max_length),
    LSTM(64),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
model.fit(
    padded_sequences,
    labels,
    epochs=10,
    batch_size=2
)
```

```python
test_texts = [
    "I really love this film",
    "This is the worst movie"
]

test_sequences = tokenizer.texts_to_sequences(test_texts)
test_padded = pad_sequences(test_sequences, maxlen=max_length, padding='post')

predictions = model.predict(test_padded)

for text, pred in zip(test_texts, predictions):
    sentiment = "Positive" if pred > 0.5 else "Negative"
    print(f"{text} --> {sentiment}")
```
